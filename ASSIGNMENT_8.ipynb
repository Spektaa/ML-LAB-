```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import pandas as pd
```

```python
!wget https://www.philippe-fournier-viger.com/spmf/datasets/retail.txt
```

```python
# Read the dataset
with open('retail.txt', 'r') as file:
    dataset = []
    for line in file:
        transaction = line.strip().split(' ')
        dataset.append(transaction)
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Define a list of support counts
support_counts = [300, 400, 500]  # Adjust these values as needed

# Apply Apriori for each support count
for support_count in support_counts:
    min_support = support_count / len(dataset)  # Calculate minimum support
    frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)
    print(f"\nFrequent Itemsets with Support Count = {support_count}:\n")
    display(frequent_itemsets)
```
