```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
```

```python
!wget https://www.philippe-fournier-viger.com/spmf/datasets/retail.txt
```

```python
# Read the dataset
with open('retail.txt', 'r') as file:
    dataset = []
    for line in file:
        transaction = line.strip().split(' ')
        dataset.append(transaction)
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Set minimum support for Apriori (adjust as needed)
min_support = 0.003  # Example value

frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)
```

```python
# Define a list of minimum confidence values
confidence_values = [0.5, 0.6, 0.7]  # Adjust these values

# Generate rules for each confidence value
for confidence in confidence_values:
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=confidence)
    print(f"\nAssociation Rules with Confidence >= {confidence}:\n")
    display(rules)
```
