Part II: Data Preprocessing and Transformation

Depending on the data type of each attribute, transform each object from your preprocessed data.

```python
# Fill null values with the median for numeric columns and the mode for categorical columns
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
    elif df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)


```

```python
# Handling missing values
from sklearn.impute import SimpleImputer

# Impute missing values with mean for numerical attributes
imputer = SimpleImputer(strategy='mean')
df_numeric = df.select_dtypes(include=['float64', 'int64'])
df_numeric_imputed = pd.DataFrame(imputer.fit_transform(df_numeric), columns=df_numeric.columns)

# For categorical attributes, use the most frequent value
imputer = SimpleImputer(strategy='most_frequent')
df_categorical = df.select_dtypes(include=['object'])
df_categorical_imputed = pd.DataFrame(imputer.fit_transform(df_categorical), columns=df_categorical.columns)

# Combine the imputed dataframes back into the original dataframe
df_imputed = pd.concat([df_numeric_imputed, df_categorical_imputed], axis=1)
```

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df_numeric_imputed), columns=df_numeric_imputed.columns)
```

```python
# Discretization (Binning) of Continuous Attributes
from sklearn.preprocessing import KBinsDiscretizer
binner = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')
df_binned = pd.DataFrame(binner.fit_transform(df_numeric_imputed), columns=df_numeric_imputed.columns)
```

```python
# Handling Categorical Attributes with Too Many Unique Values

# Frequency Encoding for Categorical Attributes
for col in df_categorical_imputed.columns:
    freq_encoding = df_categorical_imputed[col].value_counts() / len(df_categorical_imputed)
    df_categorical_imputed[col] = df_categorical_imputed[col].map(freq_encoding)

# Group Rare Categories
threshold = 0.01
for col in df_categorical_imputed.columns:
    value_counts = df_categorical_imputed[col].value_counts(normalize=True)
    rare_categories = value_counts[value_counts < threshold].index
    df_categorical_imputed[col] = df_categorical_imputed[col].replace(rare_categories, 'Rare')
```

```python
# One-Hot Encoding for Categorical Attributes
df_onehot_encoded = pd.get_dummies(df_categorical_imputed)
```

```python
# Standardization
scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df_numeric_imputed), columns=df_numeric_imputed.columns)
```

```python
# Combine all transformations into the final DataFrame
df_final = pd.concat([df_standardized, df_binned, df_onehot_encoded], axis=1)
```

```python
from geopy.geocoders import Nominatim
import time

geolocator = Nominatim(user_agent="geoapiExercises")

def get_lat_long(address_line, postal_code):
    # Added a delay to avoid rate limiting
    time.sleep(1)

    location = geolocator.geocode(f"{address_line}, {postal_code}", timeout=10)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Example usage:
df['Home Latitude'], df['Home Longitude'] = zip(*df.apply(lambda row: get_lat_long(row['AddressLine1'], row['PostalCode']), axis=1))
df['Work Latitude'], df['Work Longitude'] = zip(*df.apply(lambda row: get_lat_long(row['AddressLine1'], row['PostalCode']), axis=1))


def haversine(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    r = 6371  # Radius of Earth in kilometers. Use 3956 for miles.
    return c * r

# Calculate the commute distance
df['Commute Distance'] = df.apply(lambda row: haversine(row['Home Latitude'], row['Home Longitude'],
                                                        row['Work Latitude'], row['Work Longitude']), axis=1)
```

```python
# Correlation between Commute Distance and Yearly Income
# Pearson Correlation
pearson_corr = df['Commute Distance'].corr(df['Yearly Income'])

# Calculate Spearman Rank Correlation (useful if the relationship is monotonic but not linear)
spearman_corr = df['Commute Distance'].corr(df['Yearly Income'], method='spearman')

print(f"Pearson Correlation: {pearson_corr}")
print(f"Spearman Correlation: {spearman_corr}")
```
